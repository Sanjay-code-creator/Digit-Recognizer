```python
import keras
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
```

```python
train["label"]
```

```python
X_train=train.drop(labels=["label"],axis=1)
X_train
```

```python
y_train=train["label"]
```

```python
y_train
```

```python
X_train=X_train.astype("float32")
test=test.astype("float32")
X_train=X_train/255
test=test/255
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_train.shape
```

```python
X_train,X_train.shape
```

```python
X_train=X_train.values.reshape(-1,28,28,1)
test=test.values.reshape(-1,28,28,1)
```

```python
X_train.shape
```

```python
test.shape
```

```python
X_train,X_test,y_train,y_test= tts(X_train,y_train,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Dropout
```

```python
cf=Sequential()
```

```python
cf.add(Conv2D(32,(3,3),padding='same',input_shape=(28,28,1),activation='relu'))
cf.add(Conv2D(32,(3,3),activation='relu'))
cf.add(MaxPooling2D(pool_size=(2,2)))
cf.add(Dropout(0.25))
```

```python
cf.add(Conv2D(32,(3,3),padding='same',activation='relu'))
cf.add(Conv2D(32,(3,3),activation='relu'))
cf.add(MaxPooling2D(pool_size=(2,2)))
cf.add(Dropout(0.25))
```

```python
cf.add(Flatten())
cf.add(Dense(512,activation='relu'))
cf.add(Dropout(0.5))
cf.add(Dense(10,activation='softmax'))
```

```python
cf.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
cf.fit(X_train,y_train,batch_size=32,epochs=30,validation_data=(X_test,y_test),shuffle=True)
```

```python
results=cf.predict(test)
```

```python
test[0]
```

```python
results = np.argmax(results,axis = 1)
results = pd.Series(results,name="Label")
```

```python
submission = pd.concat([pd.Series(range(1,28001),name = "ImageId"),results],axis = 1)
submission.to_csv("C:\\Users\\sanju\\OneDrive\\Documents\\submission_Dig.csv",index=False)
```
